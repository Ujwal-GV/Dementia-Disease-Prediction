```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
from sklearn.preprocessing import  LabelEncoder

data = pd.read_csv('C:\\Users\\Lenovo\\Desktop\\FINAL_INTERNSHIP\\DEMENTIA\\dementia_project\\dementia_dataset.csv')

print(data)
print(data.info())

le_gr = LabelEncoder()
data['new_gr'] = le_gr.fit_transform(data['Group'])

median_ses = data['SES'].median()
data['SES'] = data['SES'].fillna(median_ses)

median_mmse = data['MMSE'].median()
data['MMSE'] = data['MMSE'].fillna(median_mmse)

features = data[['MR Delay', 'Age', 'EDUC', 'SES', 'MMSE', 'CDR', 'eTIV', 'nWBV', 'ASF']]
target = data['Group']

print(data)
print(data.info())
    
model = DecisionTreeClassifier()
model.fit(features, target)
    
mr = float(input("ENTER MR"))
age = float(input("ENTER AGE"))
educ = float(input("ENTER EDUC"))
ses = float(input("ENTER SES"))
mmse = float(input("ENTER MMSE"))
cdr = float(input("ENTER CDR"))
etiv = float(input("ENTER ETIV"))
nwbv = float(input("ENTER NWBV"))
asf = float(input("ENTER ASF"))
    
prediction = model.predict([[mr, age, educ, ses, mmse, cdr, etiv, nwbv, asf]])

print(prediction[0])
```
