```python
from django.shortcuts import render
from sklearn.tree import DecisionTreeClassifier
import pandas as pd
from sklearn.preprocessing import LabelEncoder
import matplotlib.pyplot as plt
import io
import urllib, base64

def predict_result(request):
    if request.method == 'POST':
        # Extracting user input
        mr = float(request.POST.get('mr'))
        age = float(request.POST.get('age'))
        educ = float(request.POST.get('educ'))
        ses = float(request.POST.get('ses'))
        mmse = float(request.POST.get('mmse'))
        cdr = float(request.POST.get('cdr'))
        etiv = float(request.POST.get('etiv'))
        nwbv = float(request.POST.get('nwbv'))
        asf = float(request.POST.get('asf'))

        # Loading dataset
        data = pd.read_csv('C:/Users/Lenovo/Desktop/FINAL_INTERNSHIP/DEMENTIA/dementia_project/dementia_dataset.csv')
        
        # Preprocessing
        le_gr = LabelEncoder()
        data['new_gr'] = le_gr.fit_transform(data['Group'])

        median_ses = data['SES'].median()
        data['SES'] = data['SES'].fillna(median_ses)

        median_mmse = data['MMSE'].median()
        data['MMSE'] = data['MMSE'].fillna(median_mmse)

        # Model training
        inputs = data[['MR Delay', 'Age', 'EDUC', 'SES', 'MMSE', 'CDR', 'eTIV', 'nWBV', 'ASF']]
        output = data['Group']

        model = DecisionTreeClassifier()
        model.fit(inputs, output)

        # Making prediction
        prediction = model.predict([[mr, age, educ, ses, mmse, cdr, etiv, nwbv, asf]])

        # Visualization
        features = ['MR Delay', 'Age', 'EDUC', 'SES', 'MMSE', 'CDR', 'eTIV', 'nWBV', 'ASF']
        plt.figure(figsize=(10, 6))
        for feature in features:
            plt.hist(data[feature], bins=20, alpha=0.7, label=feature)
        plt.legend()
        plt.xlabel('Feature Value')
        plt.ylabel('Frequency')
        plt.title('Distribution of Features')
        # Convert plot to base64 encoding
        buffer = io.BytesIO()
        plt.savefig(buffer, format='png')
        buffer.seek(0)
        image_png = buffer.getvalue()
        buffer.close()
        graphic = base64.b64encode(image_png)
        graphic = graphic.decode('utf-8')
        plt.close()

        # Returning result along with the rendered template
        return render(request, 'user_input.html', {'prediction': prediction[0], 'graphic': graphic})
    else:
        return render(request, 'user_input.html')
```

```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
from sklearn.preprocessing import LabelEncoder
import matplotlib.pyplot as plt
import numpy as np
import io
import base64

# Load dataset
data = pd.read_csv('dementia_dataset.csv')

# Preprocessing
le_gr = LabelEncoder()
data['new_gr'] = le_gr.fit_transform(data['Group'])

median_ses = data['SES'].median()
data['SES'] = data['SES'].fillna(median_ses)

median_mmse = data['MMSE'].median()
data['MMSE'] = data['MMSE'].fillna(median_mmse)

# Model training
inputs = data[['MR Delay', 'Age', 'EDUC', 'SES', 'MMSE', 'CDR', 'eTIV', 'nWBV', 'ASF']]
output = data['Group']

model = DecisionTreeClassifier()
model.fit(inputs, output)

# Set feature names
model.feature_names = inputs.columns.tolist()

# Make prediction for a sample input
sample_input = [10, 70, 12, 2, 25, 0.5, 1400, 0.8, 1.1]
prediction = model.predict([sample_input])

# Calculate probabilities for each class
proba = model.predict_proba([sample_input])
classes = model.classes_

# Plot probability distribution
plt.figure(figsize=(8, 6))
plt.bar(classes, proba[0])
plt.xlabel('Classes')
plt.ylabel('Probability')
plt.title('Probability Distribution of Classes')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Plot distribution of features
features = ['MR Delay', 'Age', 'EDUC', 'SES', 'MMSE', 'CDR', 'eTIV', 'nWBV', 'ASF']
plt.figure(figsize=(10, 6))
for feature in features:
    plt.hist(data[feature], bins=20, alpha=0.7, label=feature)
plt.legend()
plt.xlabel('Feature Value')
plt.ylabel('Frequency')
plt.title('Distribution of Features')
plt.show()
```

```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.preprocessing import LabelEncoder
import matplotlib.pyplot as plt
import numpy as np
import io
import base64

# Load dataset
data = pd.read_csv('dementia_dataset.csv')

# Preprocessing
le_gr = LabelEncoder()
data['new_gr'] = le_gr.fit_transform(data['Group'])

median_ses = data['SES'].median()
data['SES'] = data['SES'].fillna(median_ses)

median_mmse = data['MMSE'].median()
data['MMSE'] = data['MMSE'].fillna(median_mmse)

# Model training
inputs = data[['MR Delay', 'Age', 'EDUC', 'SES', 'MMSE', 'CDR', 'eTIV', 'nWBV', 'ASF']]
output = data['Group']

model = DecisionTreeClassifier()
model.fit(inputs, output)

# Set feature names
model.feature_names = inputs.columns.tolist()

# Make prediction for a sample input
sample_input = [10, 70, 12, 2, 25, 0.5, 1400, 0.8, 1.1]
prediction = model.predict([sample_input])

# Calculate probabilities for each class
proba = model.predict_proba([sample_input])
classes = model.classes_

# Plot probability distribution
plt.figure(figsize=(8, 6))
plt.bar(classes, proba[0])
plt.xlabel('Classes')
plt.ylabel('Probability')
plt.title('Probability Distribution of Classes')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Plot distribution of features
features = ['MR Delay', 'Age', 'EDUC', 'SES', 'MMSE', 'CDR', 'eTIV', 'nWBV', 'ASF']
plt.figure(figsize=(10, 6))
for feature in features:
    plt.hist(data[feature], bins=20, alpha=0.7, label=feature)
plt.legend()
plt.xlabel('Feature Value')
plt.ylabel('Frequency')
plt.title('Distribution of Features')
plt.show()

# Plot decision tree
plt.figure(figsize=(200, 100))
plot_tree(model, feature_names=model.feature_names, class_names=model.classes_, filled=True)
plt.show()

# Convert plot to base64 encoding
buffer = io.BytesIO()
plt.savefig(buffer, format='png')
buffer.seek(0)
image_png = buffer.getvalue()
buffer.close()
graphic = base64.b64encode(image_png)
graphic = graphic.decode('utf-8')
plt.close()

# Now you can pass the 'prediction' and 'graphic' variables to your template for rendering
```
